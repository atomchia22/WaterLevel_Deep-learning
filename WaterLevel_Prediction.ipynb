# LSTM

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error, r2_score
from datetime import timedelta

# --- Load CSV ---
file_path = "/content/drive/MyDrive/LBY#2_WaterLevel/DWLR_Dataset_2023.csv"
df = pd.read_csv(file_path)

# Ensure Date is datetime
df["Date"] = pd.to_datetime(df["Date"])

# Drop NaNs in Water_Level_m
data = df[["Date", "Water_Level_m"]].dropna()
dates = data["Date"].values
water_level = data["Water_Level_m"].values.reshape(-1, 1)

# Normalize water level
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(water_level)

# --- Create sequences for LSTM ---
seq_length = 10
def create_sequences(data, seq_length=10):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

X, y = create_sequences(data_scaled, seq_length)
dates_seq = dates[seq_length:]  # corresponding dates for y

# --- Train-validation split ---
X_train, X_val, y_train, y_val, dates_train, dates_val = train_test_split(
    X, y, dates_seq, test_size=0.2, shuffle=False
)

# --- Build LSTM model ---
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(seq_length, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# --- Train model ---
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=50,
    batch_size=16,
    verbose=1
)

# --- Predict on validation set ---
y_pred_scaled = model.predict(X_val)
y_pred = scaler.inverse_transform(y_pred_scaled)
y_val_original = scaler.inverse_transform(y_val)

# --- Compute Metrics ---
rmse = np.sqrt(mean_squared_error(y_val_original, y_pred))
r2 = r2_score(y_val_original, y_pred)
nse = 1 - (np.sum((y_val_original - y_pred)**2) / np.sum((y_val_original - np.mean(y_val_original))**2))

print(f"Validation Metrics:\nR2: {r2:.4f}\nRMSE: {rmse:.4f}\nNSE: {nse:.4f}")

# --- Plot training and validation loss ---
plt.figure(figsize=(8,5))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE)')
plt.title('LSTM Training & Validation Loss')
plt.legend()
plt.show()

# --- Plot predicted vs actual over time ---
plt.figure(figsize=(12,6))
plt.plot(dates_val, y_val_original, label='Actual Water Level')
plt.plot(dates_val, y_pred, label='Predicted Water Level')
plt.xlabel('Date')
plt.ylabel('Water Level (m)')
plt.title('LSTM Water Level Prediction Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# --- Plot validation metrics as bar chart ---
metrics = [rmse, r2, nse]
metric_names = ['RMSE', 'R2', 'NSE']

plt.figure(figsize=(8,5))
bars = plt.bar(metric_names, metrics, color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Validation Metrics')
plt.ylabel('Metric Value')
plt.ylim(0, max(1.1, max(metrics)*1.1))
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.02, f"{yval:.3f}", ha='center')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# --- Forecast next month (30 days) ---
forecast_days = 30
last_sequence = data_scaled[-seq_length:]  # last seq_length points
forecast_scaled = []

for _ in range(forecast_days):
    input_seq = last_sequence.reshape(1, seq_length, 1)
    pred = model.predict(input_seq)[0,0]
    forecast_scaled.append(pred)
    last_sequence = np.append(last_sequence[1:], pred).reshape(-1,1)  # slide window

forecast = scaler.inverse_transform(np.array(forecast_scaled).reshape(-1,1))
last_date = data["Date"].iloc[-1]
forecast_dates = [last_date + timedelta(days=i+1) for i in range(forecast_days)]

# --- Plot forecast ---
plt.figure(figsize=(12,6))
plt.plot(data["Date"], data["Water_Level_m"], label='Historical Water Level')
plt.plot(forecast_dates, forecast, label='Forecast Next Month', color='red')
plt.xlabel('Date')
plt.ylabel('Water Level (m)')
plt.title('LSTM Water Level Forecast for Next 30 Days')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# --- Display forecast values ---
forecast_df = pd.DataFrame({"Date": forecast_dates, "Forecast_Water_Level_m": forecast.flatten()})
print(forecast_df)
```

# ANN

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import mean_squared_error, r2_score
from datetime import timedelta

# --- Load CSV ---
file_path = "/content/drive/MyDrive/LBY#2_WaterLevel/DWLR_Dataset_2023.csv"
df = pd.read_csv(file_path)

# Ensure Date is datetime
df["Date"] = pd.to_datetime(df["Date"])

# Drop NaNs in Water_Level_m
data = df[["Date", "Water_Level_m"]].dropna()
dates = data["Date"].values
water_level = data["Water_Level_m"].values.reshape(-1, 1)

# Normalize water level
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(water_level)

# --- Create sequences for ANN ---
def create_sequences(data, seq_length=10):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length].flatten())  # flatten for ANN input
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

seq_length = 10
X, y = create_sequences(data_scaled, seq_length)
dates_seq = dates[seq_length:]  # corresponding dates for y

# --- Train-validation split ---
X_train, X_val, y_train, y_val, dates_train, dates_val = train_test_split(
    X, y, dates_seq, test_size=0.2, shuffle=False
)

# --- Build ANN model ---
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=seq_length))
model.add(Dense(32, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# --- Train model ---
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=50,
    batch_size=16,
    verbose=1
)

# --- Predict on validation set ---
y_pred_scaled = model.predict(X_val)
y_pred = scaler.inverse_transform(y_pred_scaled)
y_val_original = scaler.inverse_transform(y_val)

# --- Compute Metrics ---
rmse = np.sqrt(mean_squared_error(y_val_original, y_pred))
r2 = r2_score(y_val_original, y_pred)
nse = 1 - (np.sum((y_val_original - y_pred)**2) / np.sum((y_val_original - np.mean(y_val_original))**2))

print(f"Validation Metrics:\nR2: {r2:.4f}\nRMSE: {rmse:.4f}\nNSE: {nse:.4f}")

# --- Plot training and validation loss ---
plt.figure(figsize=(8,5))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE)')
plt.title('ANN Training & Validation Loss')
plt.legend()
plt.show()

# --- Plot predicted vs actual over time ---
plt.figure(figsize=(12,6))
plt.plot(dates_val, y_val_original, label='Actual Water Level')
plt.plot(dates_val, y_pred, label='Predicted Water Level')
plt.xlabel('Date')
plt.ylabel('Water Level (m)')
plt.title('ANN Water Level Prediction Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# --- Plot validation metrics as bar chart ---
metrics = [rmse, r2, nse]
metric_names = ['RMSE', 'R2', 'NSE']

plt.figure(figsize=(8,5))
bars = plt.bar(metric_names, metrics, color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Validation Metrics')
plt.ylabel('Metric Value')
plt.ylim(0, max(1.1, max(metrics)*1.1))
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.02, f"{yval:.3f}", ha='center')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# --- Forecast next month (30 days) ---
forecast_days = 30
last_sequence = data_scaled[-seq_length:].flatten()
forecast_scaled = []

for _ in range(forecast_days):
    input_seq = last_sequence.reshape(1, -1)
    pred = model.predict(input_seq)[0,0]
    forecast_scaled.append(pred)
    last_sequence = np.append(last_sequence[1:], pred)  # slide window

forecast = scaler.inverse_transform(np.array(forecast_scaled).reshape(-1,1))
last_date = data["Date"].iloc[-1]
forecast_dates = [last_date + timedelta(days=i+1) for i in range(forecast_days)]

# --- Plot forecast ---
plt.figure(figsize=(12,6))
plt.plot(data["Date"], data["Water_Level_m"], label='Historical Water Level')
plt.plot(forecast_dates, forecast, label='Forecast Next Month', color='red')
plt.xlabel('Date')
plt.ylabel('Water Level (m)')
plt.title('ANN Water Level Forecast for Next 30 Days')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# --- Display forecast values ---
forecast_df = pd.DataFrame({"Date": forecast_dates, "Forecast_Water_Level_m": forecast.flatten()})
print(forecast_df)
```
